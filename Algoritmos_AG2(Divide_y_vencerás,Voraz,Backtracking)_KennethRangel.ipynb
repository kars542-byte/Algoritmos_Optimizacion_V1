```python
#Torres de Hanoi - Divide y venceras
################################################################################
import time
import matplotlib.pyplot as plt

################################################################
def Torres_Hanoi(N, desde, hasta):
  #N - Nº de fichas
  #desde - torre inicial
  #hasta - torre fina
  if N==1 :
    print("Lleva la ficha desde " + str(desde) + " hasta " + str(hasta))

  else:
    Torres_Hanoi(N-1, desde, 6-desde-hasta)
    print("Lleva la ficha desde " + str(desde) + " hasta " + str(hasta))
    Torres_Hanoi(N-1, 6-desde-hasta,  hasta)
comparador = [0,1,2]
start = time.time()
Torres_Hanoi(4, 2, 3)
end = time.time()
comparador[0] = round((end-start),5)
start = time.time()
Torres_Hanoi(10, 1, 2)
end = time.time()
comparador[1] = round((end-start),5)
start = time.time()
Torres_Hanoi(12, 3, 2)
end = time.time()
comparador[2] = round((end-start),5)
################################################################
plt.plot([4,10,12], comparador, 'ro-')
plt.xlabel('Cantidad de discos')
plt.ylabel('Tiempo (s)')
plt.title('Tiempo de ejecucion')
plt.grid(True)
plt.show()
```

#Cambio de monedas - Técnica voraz

```python
#Cambio de monedas - Técnica voraz
################################################################################
SISTEMA = [11, 5 ,1  ]
SISTEMA2 = [1000, 100, 50, 10, 1]
SISTEMA3 = [10000, 5000, 2000, 1000, 100, 500, 200, 100, 50, 20, 10, 5, 2, 1]
################################################################
def cambio_monedas(CANTIDAD,SISTEMA):
#....
  SOLUCION = [0]*len(SISTEMA)
  ValorAcumulado = 0

  for i,valor in enumerate(SISTEMA):
    monedas = (CANTIDAD-ValorAcumulado)//valor
    SOLUCION[i] = monedas
    ValorAcumulado = ValorAcumulado + monedas*valor

    if CANTIDAD == ValorAcumulado:
      return SOLUCION

  print("No es posible encontrar solucion")

comparador = [0,1,2]
start = time.time()
print(cambio_monedas(15,SISTEMA))
end = time.time()
comparador[0] = round((end-start),5)
start = time.time()
print(cambio_monedas(2569,SISTEMA2))
end = time.time()
comparador[1] = round((end-start),5)
start = time.time()
print(cambio_monedas(256065,SISTEMA3))
end = time.time()
comparador[2] = round((end-start),5)
################################################################
plt.plot([3,5,10], comparador, 'ro-')
plt.xlabel('Cantidad de discos')
plt.ylabel('Tiempo (s)')
plt.title('Tiempo de ejecucion')
plt.grid(True)
plt.show()
```

#N Reinas - Vuelta Atrás(Backtracking)

```python
#N Reinas - Vuelta Atrás()
################################################################################

#Verifica que en la solución parcial no hay amenzas entre reinas
################################################################
def es_prometedora(SOLUCION,etapa):
################################################################
  #print(SOLUCION)
  #Si la solución tiene dos valores iguales no es valida => Dos reinas en la misma fila
  for i in range(etapa+1):
    #print("El valor " + str(SOLUCION[i]) + " está " +  str(SOLUCION.count(SOLUCION[i])) + " veces")
    if SOLUCION.count(SOLUCION[i]) > 1:
      return False

    #Verifica las diagonales
    for j in range(i+1, etapa +1 ):
      #print("Comprobando diagonal de " + str(i) + " y " + str(j))
      if abs(i-j) == abs(SOLUCION[i]-SOLUCION[j]) : return False
  return True

#Traduce la solución al tablero
################################################################
def escribe_solucion(S):
################################################################
  n = len(S)
  for x in range(n):
    print("")
    for i in range(n):
      if S[i] == x+1:
        print(" X " , end="")
      else:
        print(" - ", end="")



#Proceso principal de N-Reinas
################################################################
def reinas(N, solucion=[],etapa=0):
################################################################
### ....
  if len(solucion) == 0:         # [0,0,0...]
    solucion = [0 for i in range(N) ]

  for i in range(1, N+1):
    solucion[etapa] = i
    if es_prometedora(solucion, etapa):
      if etapa == N-1:
        print(solucion)
      else:
        reinas(N, solucion, etapa+1)
    else:
      None

  solucion[etapa] = 0

comparador = [0,1,2]
start = time.time()
reinas(5,solucion=[],etapa=0)
end = time.time()
comparador[0] = round((end-start),5)
start = time.time()
reinas(7,solucion=[],etapa=0)
end = time.time()
comparador[1] = round((end-start),5)
start = time.time()
reinas(10,solucion=[],etapa=0)
end = time.time()
comparador[2] = round((end-start),5)
################################################################
plt.plot([5,7,10], comparador, 'ro-')
plt.xlabel('Cantidad de discos')
plt.ylabel('Tiempo (s)')
plt.title('Tiempo de ejecucion')
plt.grid(True)
plt.show()
```
